# 엑셀파일 데이터프레임으로 받기

```python
import pandas as pd
```

```python
df = pd.read_csv('./tickers.csv')
```

```python
df1 = df[2:]
```

```python
df2 = df1.rename(columns=df1.iloc[0])
```

```python
df3 = df2[1:]
```

```python
df4= df3.reset_index();df4
```

```python
df4.isnull().sum()
```

```python
len(df4)
```

```python
del df4['index']
```

```python
df5 = df4.dropna(how = 'all',axis = 1)
```

```python
df5
```

# 시장 별로 데이터프레임 만들기

```python
gdf = df5.groupby(df5['Exchange']).count()
```

```python
exchanges = gdf.index
```

```python
a = pd.DataFrame()
a
```

```python
for i in exchanges:
    i=+1
    x = df5.loc[df5['Exchange']== exchanges[i]]
    pd.concat(a,x)
        
    
```
