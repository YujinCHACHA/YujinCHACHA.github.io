```python
df = pd.read_csv('/content/drive/MyDrive/DNA/TITANIC/train.csv')
df
```

1) 사망자와 생존자의 비율을 원 그래프로 나타내기

```python
x = df['Survived'].value_counts().index
y = df['Survived'].value_counts().values
plt.figure(dpi = 150)

plt.title('사망자 vs 생존자')

plt.pie(y,labels = x)
plt.legend()   # 이 위치에 있어야함
plt.show()
```

2) 객실 등급의 비율을 원 그래프로 나타내기

```python
x = df['Pclass'].value_counts().index
y = df['Pclass'].value_counts().values
plt.figure(dpi = 150)

plt.title('객실등급 비율')

plt.pie(y,labels = x)
plt.legend()
plt.show()
```

3) 객실 등급 별 객실 수를 막대그래프로 나타내기

```python
df['Pclass'].value_counts().sort_index()
```

```python
y = df['Pclass'].value_counts().values
plt.figure(dpi = 150)

plt.title('객실등급 비율')
plt.xlabel('객실등급')
plt.ylabel('객실 수')
plt.xticks([1,2,3])

plt.bar(x,y)
plt.show()
```

```python
y = df['Pclass'].value_counts().values
plt.figure(dpi = 150)

plt.title('객실등급 비율')
plt.xlabel('객실등급')
plt.ylabel('객실 수')
plt.xticks([1,2,3])

plt.bar(x,y)
plt.show()
```

4) 객실 등급 별 가격

```python
plt.figure(dpi=150)

x = df.groupby('Pclass').mean().sort_values(by='Fare', ascending = False).index
y = df.groupby('Pclass').mean().sort_values(by='Fare', ascending = False)['Fare']

plt.title('객실 등급 별 가격')
plt.xlabel('객실 등급')
plt.xticks([1,2,3])
plt.ylabel('가격')
plt.bar(x,y)
plt.show()
```
