- 컬럼명: Code, Date, Close, Open, High, Low, Volume, Change
- 데이터타입 변경 (주가, 거래대금: int, 등락률: float)

# 1. import library

```python
import pandas as pd
```

# 2. Load Data

```python
df = pd.read_csv('KONEX_20130102_20181230.csv');df
```

# 3. Extract Columns

```python
data = df.loc[:,['Code', '일자', '종가', '시가', '고가', '저가', '거래량' , '등락률']]
```

```python
data.columns = ['Code', 'Date', 'Close', 'Open', 'High', 'Low', 'Volume', 'Change']
```

# 4. Change Data type

```python
data['Date']= data['Date'].str.replace(pat=r'"',repl=r'',regex=True)
data['Close']= data['Close'].str.replace(pat=r'"',repl=r'',regex=True)
data['Open']= data['Open'].str.replace(pat=r'"',repl=r'',regex=True)
data['High']= data['High'].str.replace(pat=r'"',repl=r'',regex=True)
data['Low']= data['Low'].str.replace(pat=r'"',repl=r'',regex=True)
data['Volume']= data['Volume'].str.replace(pat=r'"',repl=r'',regex=True)
data['Change']= data['Change'].str.replace(pat=r'"',repl=r'',regex=True)
```

```python
data
```

```python
data.dtypes
```

```python
data2 = data[['Close', 'Open', 'Close','High', 'Low', 'Volume', 'Change']].apply(pd.to_numeric)
```

```python
data2['Date'] = data['Date']
```

```python
data2.dtypes
```

```python
data2.to_csv('KONEX_20130102_20131230.csv',index=None)
```
